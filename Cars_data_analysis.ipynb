```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
Y_ommited = np.log(Y)
```

```python
print(type(Y_ommited))
```

```python
print(Y_ommited)
```

```python
#spliting data into train and test
x_train,x_test,y_train,y_test = train_test_split(X,Y_ommited,test_size=0.3,random_state=1)
```

```python
"""
We are making a base model by using test data mean value
This is to set a benchmark and to compare with our regression model
"""
#for finding mean value of test data
base_pred = np.mean(y_test)
print(type(base_pred))
print(base_pred)

# Repeating same value till length of test data
base_pred = np.repeat(base_pred, len(y_test))
print(type(base_pred))
print(base_pred)

# finding the RMSE
base_root_mean_square_error = np.sqrt(mean_squared_error(y_test,base_pred))
print(type(base_root_mean_square_error))
print(base_root_mean_square_error)
```

Building Linear Rregression Model with ommited data

```python
lgr = LinearRegression(fit_intercept=True)
```

```python
# Model
lin_model = lgr.fit(x_train,y_train)
```

```python
#prediction on linear regression
lin_regr_cars = lgr.predict(x_test)
```

```python
# Computing MSE and RMSE
lin_mse = mean_squared_error(y_test,lin_regr_cars)
lin_rmse = np.sqrt(lin_mse)
print(type(lin_mse))
print(type(lin_rmse))
print(lin_mse)
print(lin_rmse)
```

```python
# R squared value
r2_lin_test = lin_model.score(x_test,y_test)
r2_lin_train = lin_model.score(x_train,y_train)
print(type(r2_lin_train))
print(r2_lin_train,r2_lin_train)
```

```python
# Regression diagnostics- Residual plot analysis
residuals = y_test - lin_regr_cars
sns.regplot(x=lin_regr_cars, y=residuals, scatter=True, fit_reg=False)
residuals.dtype
```

Random Forest model on ommited data

```python
# model building
random_forest = RandomForestRegressor(n_estimators = 100, max_features = 'auto', max_depth = 100, min_samples_split = 10, min_samples_leaf = 4, random_state = 1)
```

```python
random_forest_model = random_forest.fit(x_train,y_train)
```

```python
random_forest_cars = random_forest.predict(x_test)
```

```python
# Computing MSE and RMSE
random_forest_mse = mean_squared_error(y_test,random_forest_cars)
random_forest_rmse = np.sqrt(random_forest_mse)
print(random_forest_rmse)
```

```python
# R squared value
r2_random_forest_test = random_forest_model.score(x_test,y_test)
r2_random_forest_train = random_forest_model.score(x_train,y_train)
print(r2_random_forest_test,r2_random_forest_train)
```
